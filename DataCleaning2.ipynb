```python
import pandas as pd
import json
```

# Data Cleaning

```python
def extract_info(row):
    parts = row.split(' · ')
    description = parts[0]
    stars = ''
    bedrooms = ''
    beds = ''
    bathrooms = ''
    
    for part in parts[1:]:
        if '★' in part:
            stars = part
            stars = part.replace('★', '')
        elif 'bedroom' in part:
            bedrooms = part
        elif 'beds' in part or 'bed' in part:
            beds = part
        elif 'bath' in part:
            bathrooms = part
    
    return pd.Series([description, stars, bedrooms, beds, bathrooms])
```

## Amsterdam

```python
amsterdam_listings = pd.read_csv('data/airbnb/amsterdam/listings.csv')
amsterdam_listings.head()
```

```python
amsterdam_listings[['description', 'stars', 'bedrooms', 'beds', 'bathrooms']] = amsterdam_listings['name'].apply(lambda x: pd.Series(extract_info(x)))
```

```python
# Count the number of null values in each column
null_counts = amsterdam_listings.isnull().sum()
print(null_counts)
```

```python
amsterdam_listings = amsterdam_listings.dropna(subset=['price'])
amsterdam_listings.drop(columns=['neighbourhood_group', 'license', 'last_review', 'reviews_per_month'], inplace=True)
```

```python
amsterdam_listings.to_csv('data/airbnb/amsterdam/listings_cleaned.csv', index=False)
```
