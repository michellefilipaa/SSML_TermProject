```python
names = pd.read_csv("/Users/ashademeij/Downloads/Spatial Statistics and Machine Learning/Project/neighbourhoods.csv")
names = names.drop(columns=['Unnamed: 0'])
names.head()
```

```python
# create dictionary old vs new
name_mapping = dict(zip(names['past_neighbourhood_158'], names['neighbourhood']))
census2021['Neighbourhoods'] = census2021['Neighbourhoods'].map(name_mapping).fillna(census2021['Neighbourhoods'])

census2021.head()
```

```python
census2021 = census2021.groupby('Neighbourhoods').sum()
census2021 = census2021.reset_index()
print(census2021.shape)
census2021.head()
```
